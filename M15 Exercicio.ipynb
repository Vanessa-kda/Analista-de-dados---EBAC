## 1\. Preços `outliers` de diamante

```python
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```python
# gráfico do exercício 1

diamonds =  data[['price']]

with sns.axes_style('whitegrid'):
    grafico = sns.boxplot(data=diamonds, x='price')
    grafico.set(title='Preços dos diamante')
```

**Insight do gráfico 1**: ...

É possivel notar que os principais valores são entre 1mil á uma média de R$ 5.100.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```python
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```python
diamonds = data[['carat', 'clarity', 'price']]
```

```python
# gráfico do exercício 2

with sns.axes_style("ticks"):

  grafico = sns.scatterplot(data=diamonds, x="carat", y="price", hue="clarity", palette="pastel")
  grafico.set(title="Correlação dos diamantes ", xlabel="Peso", ylabel="Preço")
  grafico.get_legend().set_title("Tranparência")
```

**Insight do gráfico 2**: ...

Praticamente todas as claridades apresentam crescimento parecido, com exeção da I1, que está mais distante da curva de crescimento

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros
