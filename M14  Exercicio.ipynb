## 1\. Preço do diamante por tipo de corte

```python
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```python
# cut / corte price/preço 
```

```python
# gráfico do exercício 1

diamonds = sns.barplot(data=data, y='price', x='cut', ci=None, palette="pastel")
diamonds.set(title= 'Preço do diamante por tipo de corte', xlabel= 'Tipos', ylabel= 'Preço' )
```

**Insight do gráfico 1**: 

Os valores dos cortes estão acima de 3mil, mas não existe tanta diferença entre eles.

O Corte Premiun tem o maior valor, e o Ideal o menor valor.

## 2\. Número de passageiros em dezembro por ano

```python
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```python
# gráfico do exercício 2
#filtrar apenas os dados referentes ao mes de Dezembro
flights = data.loc[data['month']== 'Dec', ['year', 'passengers']].reset_index(drop=True)
flights

```

```python
#Para colocar grades no grafico de linha, basta incluir o With
with sns.axes_style ('whitegrid'):
    grafico= sns.lineplot(data=flights, x='year', y='passengers', palette="pastel")
    grafico.set(title='Número de passageiros em dezembro por ano', xlabel='Ano', ylabel='Passageiros');
```

**Insight do gráfico 2**: ...

Em nenhum ano, houve queda no número de passageiros, só aumento.

Entre 1957 e 1958, o numero de passageiros de manteve na média.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```python
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```python
# gráfico do exercício 3

flights = data.query("1949 <= year < 1960") #Filtrar os anos 

with sns.axes_style('whitegrid'):
  grafico_2 = sns.lineplot(data=flights, x="month", y="passengers", hue="year", palette="flare")
  grafico_2.set(title='Número de Passageiros por Mês entre 1949 e 1959', xlabel='Mês', ylabel='Passageiros');
  grafico_2.get_legend().set_title("Ano");
     
```

**Insight do gráfico 3**: ...

Os meses entre junho e agosto são os que tem maior numero de passageiros, todos os anos.

A Cada ano que se passa os numeros de passageiros vem aumentando.
